# Leyendo archivos con Python

```python
INPUT_FILE_PATH = '/content/drive/MyDrive/Datasets/datalake/datalake_inputs/netflix_titles.csv'
OUTUT_FILE_PATH_PY = '/content/drive/MyDrive/Datasets/datalake/datalake_outputs/netflix_titles_py.csv'
```

```python
# Lectura linea x linea
with open(INPUT_FILE_PATH, 'r', encoding='utf-8') as f:
  for linea in f:
    print(linea.strip())
    break
```

```python
# Lectura todas las lineas (Como lista)
with open(INPUT_FILE_PATH, 'r', encoding='utf-8') as f:
  lineas = f.readlines()

lineas[:3]
```

```python
with open(OUTUT_FILE_PATH_PY, 'w', encoding='utf-8') as f:
  for linea in lineas[:3]:
    f.write(linea)
```
