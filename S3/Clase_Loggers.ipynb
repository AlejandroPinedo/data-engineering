```python
import logging
import os
import time
```

```python
# Zona horaria de Perú (Colab <Linux>)
os.environ['TZ'] = 'America/Lima' # Actualiza variable de entorno
time.tzset()  # Aplica el cambio al proceso actual
```

```python
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

# Definimos un nombre para el logger
logger = logging.getLogger("ETL")

# Definimos el tipo (level) del mensaje a mostrar
# Nivel de logging: DEBUG --> INFO --> WARNING --> ERROR --> CRITICAL
# Si se elige un nivel, todos los niveles de la derecha se muestran
logger.setLevel(logging.DEBUG)

# Python muestra un mensaje por cada handler
# --> Eliminando en caso se haya definido más de uno

if logger.hasHandlers():
    logger.handlers.clear()

# --- Formato ---
formatter = logging.Formatter(
    "%(asctime)s - %(name)s - %(levelname)s - %(message)s"
)

# Configurando el handler
# --- Consola ---
console_handler = logging.StreamHandler()
console_handler.setFormatter(formatter)
logger.addHandler(console_handler)

# --- Archivo ---
file_handler = logging.FileHandler("etl.log", mode = "a") # w for overwrite
file_handler.setFormatter(formatter)
logger.addHandler(file_handler)
```

```python
def main():
  # Nivel de logging: DEBUG --> INFO --> WARNING --> ERROR --> CRITICAL
  logger.debug('Mensaje modo DEBUG')
  logger.info('Mensaje modo INFO')
  logger.warning('Mensaje modo WARNING')
  logger.error('Mensaje modo ERROR')
  logger.critical('Mensaje modo CRITICAL')
```

```python
logger.setLevel(logging.INFO)
```

```python
if __name__ == 'main':
    main()
```

```python
main()
```
